```python
def spectrumGrab():
    s=socket.socket()
    s.connect(('131.217.63.179', 5025))
    s.send(b'FREQ:SPAN 1GHZ\rn\n')
    s.send(b'FREQ:CENT 505MHZ\r\n')
    s.send(b'TRAC? TRACE2\r\n')
    time.sleep(0.5)
    a=s.recv(16384)
    a_list = a.decode("utf-8").split(',')
    float_list = [float(point) for point in a_list]
    return float_list 
```

```python
with open('../at0002/at0002_scantimes.out', 'r') as f:
    time_data = [float(line.strip()) for line in f]
    
scan_times = Time(time_data, format='jd', scale='utc')
```

```python
spec_list = []
while True:
    currenttime = Time(datetime.utcnow())
    for i in range(0, len(scan_times)-1):
        if scan_times[i+1] > currenttime > (scan_times[i] + timedelta(seconds = 90)):
            if not os.path.exists(cwd + '/' + str(scan_times[i]) + '.dat'):
                open(cwd + '/' + str(scan_times[i]) + '.dat', 'a').close()
            spectrum = spectrumGrab()
            spec_list.append(spectrum)
            time.sleep(5.0)
            current_scan = i
            break    
        elif i == len(scan_times)-2 and len(spec_list) > 0:
            spec_array = np.array(spec_list)
            spec_array_avg = spec_array.mean(axis=0)
            with open(cwd + '/' + str(scan_times[current_scan]) + '.dat', 'a') as f:
                for element in spec_array_avg:
                    print(element, file=f)
            spec_list = []
        else:
            continue
```

```python
def specGrab12():
    for i in range(0,12):
        spectrum = np.array(spectrumGrab())
        if i == 0:
            spec_array = spectrum
        else:
            spec_array = np.vstack((spec_array, spectrum))
    avg_array = spec_array.mean(axis=0)
    return avg_array
```
